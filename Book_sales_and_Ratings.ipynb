```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df =pd.read_csv("/content/sample_data/Books_Data_Clean.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df =df[df["Publishing Year"] > 1900
       ]
```

```python
df.isna().sum()
```

```python
df.dropna(subset = "Book Name", inplace = True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
import matplotlib.pyplot as plt
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in each Genre")
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending = False)
```

```python
sns.boxplot(x="genre", y ="Book_ratings_count", data = df)
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box plot of Books Rating count for each genre")
plt.show()

```

```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("sale price")
plt.ylabel("units sold")
plt.title("Scatter plot of Sale price vs units sold")
plt.show()
```

```python
language_count = df["language_code"].value_counts()
```

```python
plt.pie(language_count, labels =language_count.index, startangle=90)
plt.title("Language Distribution of Books")
plt.show()
```

```python
df.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending = False)
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending = False)
```

```python
df.groupby("language_code").size().sort_values(ascending = False)
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].var()
```

```python
plt.scatter(df["Book_average_rating"],df["Book_ratings_count"])
plt.xlabel("Book Average Rating")
plt.ylabel("Books Ratings count")
plt.title("Scatter plot of Book Average Rating Vs Books Ratings count")
plt.show()
```

```python
total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending = False).head().plot(kind ="bar")
plt.xlabel("Author")
plt.ylabel("Total Grosssales")
plt.title("Total gross sales by Author")
plt.show()
```

```python
sns.boxplot(x="Author_Rating", y ="units sold", data = df)
plt.xlabel("Author_Rating")
plt.ylabel("units sold")
plt.title("Author_Rating Vs units sold")
plt.show()
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind = "line", marker = 0)
plt.xlabel("Publishing year")
plt.ylabel("units sold")
plt.title("Total units sold over the years")
plt.show()
```
